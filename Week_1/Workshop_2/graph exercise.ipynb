```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as grid

```

```python
def func(x):
    return (x**3+2)/(x**3+6*x**2-4*x-4)
```

```python
x=np.linspace(-2,4,100)
y=func(x)-1
```

```python
plt.figure()
plt.plot(x,y)
plt.show()
```

```python
plt.figure()
plt.title('A mathematical function')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.plot(x,y, label='a curve', color='red')
plt.legend()
plt.savefig('simple_plot.pdf',dpi=300)
plt.show()
```

```python
data=np.random.randn(2,100)
fig, axs=plt.subplots(3,2,figsize=(10,15))

axs[0,0].hist(data[0])
axs[0,0].grid(True)
axs[0,0].legend('Hi')

axs[1,0].scatter(data[0],data[1], color="green", s=100)
axs[1,0].set_xlabel('Horizontal axis')
axs[1,0].set_ylabel('Vertical axis')

col_a=np.linspace(-2,2,100)
col_b=2*np.cos(col_a*10)
axs[1,0].plot(col_a,col_b, label='a random cos function', c='orange')
axs[1,0].legend()

axs[0,1].plot(data[0], data[1], 'bo', linestyle='dashed', markersize=12)
axs[0,1].title.set_text('This one even has a title!')

axs[1,1].hist2d(data[0], data[1], cmap="viridis")

plt.show()
```

```python
t=np.arange(0.0, 2.0, 0.01)
s=np.sin(2*np.pi*t)

plt.plot(t,s, color='green', linestyle='dashed')
plt.title(r'$\alpha_i > \beta_i$', fontsize=20)
plt.text(1,-0.6, r'$\sum_{i=0}^\infty x_i$', fontsize=20)
plt.text(0.6,0.6, r'$A\mathrm{sin}(2\omega t)$', fontsize=20)
plt.xlabel('time (s)')
plt.ylabel('volts (mV)')
plt.show()
```
