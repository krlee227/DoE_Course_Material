```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
np.random.seed(42)
```

```python
num_samples=100
num_features=50
X=np.random.randn(num_samples, num_features)
```

```python
pca=PCA()
X_pca=pca.fit_transform(X)
```

```python
explained_variance_ratio=pca.explained_variance_ratio_
```

```python
explained_variance=pca.explained_variance_
```

```python
num_components=len(explained_variance_ratio)
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(np.arange(1,num_components+1), explained_variance_ratio, 'bo-')
plt.xlabel('principal pomponent')
plt.ylabel('Explained Variance')
plt.title('scree Plot - Explained Viarance')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
print(iris)
```

```python
X=iris.data
print(X)
```

```python
y=iris.target
print(y)
```

```python
lda=LinearDiscriminantAnalysis(n_components=2)
X_lda=lda.fit_transform(X,y)
```

```python
target_names = iris.target_names
colors = ['navy', 'turquoise', 'darkorange']
plt.figure()
```

```python
for color, i, target_name in zip(colors, [0,1,2], target_names):
    plt.scatter(X_lda[y==i,0], X_lda[y==i, 1], color=color, lw=2, label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of iris dataset')
plt.show()
```
